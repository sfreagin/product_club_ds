##### Grabbing GA hire data
And very light text cleaning

```python
df_hiredata = pd.read_csv('./CSV_data/test_data.csv')

df_hiredata['Job Title'] = [ title.lower() for title in df_hiredata['Job Title'] ]

df_hiredata.head()
```

## Example Charts
#### All charts count the number of jobs accepted by a GA student, organized by a category of interest

```python
plt.figure(figsize=(12,4))
sns.countplot(df_hiredata['Job Employer Name'],color='g')
plt.xticks(rotation=90)
plt.title("Jobs accepted May/June 2021, by Company");
```

```python
plt.figure(figsize=(12,4))
sns.countplot(df_hiredata['Job Title'],color='g')
plt.xticks(rotation=90)
plt.title("Jobs accepted May/June 2021, by Job Title");
```

```python
plt.figure(figsize=(12,4))
sns.countplot(df_hiredata['Industry'],color='g')
plt.xticks(rotation=90)
plt.title('Jobs accepted May/June 2021, by Industry');
```

```python
plt.figure(figsize=(6,4))
sns.countplot(df_hiredata[df_hiredata['Metro'] != 'ONL']['Metro'],color='g')
plt.xticks(rotation=90)
plt.title("Jobs accepted May/June 2021, Metro area (excluding online)");
```

```python
plt.figure(figsize=(6,4))
sns.countplot(df_hiredata['Metro'],color='g')
plt.xticks(rotation=90);
plt.title("Jobs accepted May/June 2021, Metro area (inluding online)");
```
